```python
import functools
import psycopg2

def connection_handler(func):
    @functools.wraps(func)
    def inner(*args, **kwargs):
        # Connect to your PostgreSQL database
        connection = psycopg2.connect(
            dbname="your_database",
            user="your_username",
            password="your_password",
            host="your_host",
            port="your_port"
        )
        cursor = connection.cursor()

        try:
            # Call the decorated function with the cursor and other arguments
            result = func(cursor, *args, **kwargs)
            connection.commit()  # Commit changes to the database
            return result
        except Exception as e:
            connection.rollback()  # Rollback changes if an exception occurs
            raise e
        finally:
            # Close the cursor and connection
            cursor.close()
            connection.close()

    return inner

# Example of using the decorator
@connection_handler
def insert_data(cursor, data):
    # Assuming data is a dictionary with values to be inserted into the table
    sql = "INSERT INTO your_table (column1, column2, column3) VALUES (%s, %s, %s)"
    cursor.execute(sql, (data['column1'], data['column2'], data['column3']))

# Example usage
data = {'column1': 'value1', 'column2': 'value2', 'column3': 'value3'}
insert_data(data)
```
